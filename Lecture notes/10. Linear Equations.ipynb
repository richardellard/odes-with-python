**Definition.** On ODE for the unknown function $y(x)$ is called *linear* if it can be written in the form
$$
    a_0(x)y+a_1(x)y'(x)+\cdots+a_n(x)y^{(n)}=b(x),
$$
where $a_0,a_1,\ldots,a_n$ are functions of $x$.

# Linear first-order equations

If the equation
$$
    \frac{dy}{dx}=f(x,y)
$$
can be multiplied by an appropriate factor $\rho(x,y)$ such that both sides of the resulting equation can be easily recognised as a derivative with respect to $x$, then we call $\rho(x,y)$ an *integrating factor*.

In particular, consider the general linear first-order ODE
$$
    \frac{dy}{dx}+P(x)y=Q(x).\tag{1}
$$
Multiplying both sides of (1) by the integrating factor
$$
    \rho(x)=\exp\left(\int P(x)\,dx\right)
$$
gives
$$
    \frac{d}{dx}[y(x)\rho(x)]=y'(x)\rho(x)+y(x)\rho'(x)=Q(x)\rho(x)
$$
and hence we obtain the general solution
$$
    y(x)=\frac{1}{\rho(x)}\int Q(x)\rho(x)\,dx.
$$

**Remark 10.1.** When evaluating the integrating factor, there is no need to explicitly include an arbitrary constant of integration. Can you see why this is the case?

**Example 10.1.** Use an integrating factor to solve the initial value problem
$$
    \frac{dy}{dx}-y=\frac{11}{8}e^{-x/3},\hspace{10mm}y(0)=-1.
$$

# Linear second-order equations

The general linear second-order ODE can be written in the form
$$
    A(x)y''(x)+B(x)y'(x)+C(x)y(x)=F(x).\tag{2}
$$

**Example 10.2.** The equation
$$
    e^xy''+y'\cos(x)+\left(1+\sqrt{x}\right)y=\tan^{-1}x
$$
is linear, but the equations
$$
    y''=yy'
$$
and
$$
    y''+3(y')^2+4y^3=0
$$
are not.

**Definition.** If $F(x)\equiv0$ in (2), we say the linear second-order ODE is *homogeneous*.

With the general linear second-order equation (2), we associate the homogeneous equation
$$
    A(x)y''(x)+B(x)y'(x)+C(x)y(x)=0.
$$
Dividing across by $A(x)$ in (2) gives the general form
$$
    y''(x)+p(x)y'(x)+q(x)y=f(x)\tag{3}
$$
in the nonhomogeneous case and
$$
    y''(x)+p(x)y'(x)+q(x)y=0\tag{4}
$$
in the homogeneous case.

**Theorem 10.1. (Principal of Superposition for Homogeneous Equations)** If $y_1$ and $y_2$ are two solutions of the homogeneous equation (4) on the interval $I$, then so is the linear combination $c_1y_1+c_2y_2$ for any choice of constants $c_1$ and $c_2$.

The proof of Theorem 10.1 follows from the linearity of differentiation.

**Example 10.3.** Since $y_1(x)=\cos(x)$ and $y_2(x)=\sin(x)$ are both solutions of the equation $y''(x)+y(x)=0$, so is any linear combination $y(x)=c_1\cos(x)+c_2\sin(x)$. In fact, *every* solution of the equation takes this form.

**Theorem 10.2. (Existence and Uniqueness for Linear Equations)** Let $p$, $q$ and $f$ be continuous on the open interval $I$ containing the point $a$. Then, given numbers $b_0$ and $b_1$, the equation (3) has a unique solution on $I$ satisfying the initial conditions $y(a)=b_0$, $y'(a)=b_1$.

**Definition.** Two functions defined on an open interval $I$ are said to be *linearly dependent* on $I$ if one is a constant multiple of the other. Otherwise, the functions are called *linearly independent*.

**Theorem 10.3. (General Solutions of Homogeneous Equations)** Let $y_1$ and $y_2$ be two linearly independent solutions of the homogeneous equation (4), with $p$ and $q$ continuous on the open interval $I$. If $y$ is any solution of (4) on $I$, then there exist numbers $c_1$ and $c_2$ such that $y(x)=c_1y_1(x)+c_2y_2(x)$.

**Theorem 10.4. (General Solutions of Nonhomoheneous Equations)** Let $y_p$ be a particular solution of (3) on the open interval $I$ with the functions $p$, $q$ and $f$ continuous on $I$. Let $y_1$ and $y_2$ be two linearly independent solutions of the associated homogeneous equation (4). If $y$ is any solution of (3) on $I$, then there exist constants $c_1$ and $c_2$ such that $y(x)=c_1y_1(x)+c_2y_2(x)+y_p(x)$ for all $x\in I$.

## Linear second-order homogeneous equations with constant coefficients

Any linear second-order homogeneous ODE with constant coefficients can be written in the form
$$
    ay''(x)+by'(x)+cy(x)=0\tag{5}
$$
for some constants $a$, $b$ and $c$.

Let us try to find a solution to (5) of the form
$$
    y(x)=e^{rx}\tag{6}
$$
for some appropriate choice of $r$. Substituting (6) into (5) gives
$$
    e^{rx}\left[ ar^2+br+c \right]=0
$$
and hence we call the equation
$$
    ar^2+br+c=0\tag{7}
$$
the *characteristic equation* of (5).

### Characteristic equations with two distinct real roots

**Remark 10.2.** If $r_1$ and $r_2$ are two distinct real roots of (7), then $y_1(x)=e^{r_1x}$ and $y_2(x)=e^{r_2x}$ are two linearly independent solutions of (5).

**Example 10.4.** Find the general solution of the equation $2y''(x)-7y'(x)+3y(x)=0$.

### Characteristic equations with a repeated root

**Remark 10.3.** If $r_0$ is a repeated root of (7), that is, $ar^2+br+c=a(r-r_0)^2$, then $y_1(x)=e^{r_0x}$ and $y_2(x)=xe^{r_0x}$ are two linearly independent solutions of (5).

**Example 10.5.** Solve the initial value problem $y''(x)+2y'(x)+y=0$, $y(0)=5$, $y'(0)=-3$.

### Characteristic equations with complex roots

Now suppose the roots of (7) are given by $r_1=\alpha+i\beta$ and $r_2=\alpha-i\beta$, where $\alpha\in\mathbb{R}$ and $\beta>0$. Then the general solution is given by
\begin{align*}
    y(x)&=C_1e^{(\alpha+i\beta)x}+C_2e^{(\alpha-i\beta)x}\\
    &=e^{\alpha x}\left[ C_1e^{i\beta x}+C_2e^{-i\beta x} \right]\\
    &=e^{\alpha x}\left[ C_1\left(\cos(\beta x)+i\sin(\beta x)\right)+C_2\left(\cos(\beta x)-i\sin(\beta x)\right) \right]\\
    &=e^{\alpha x}\left[ (C_1+C_2)\cos(\beta x) + i(C_1-C_2)\sin(\beta x) \right]\\
    &=c_1e^{\alpha x}\cos(\beta x)+c_2e^{\alpha x}\sin(\beta x),
\end{align*}
where
$$
    c_1=C_1+C_2, \hspace{10mm}c_2=i(C_1-C_2).
$$

Observe that, since $y$ is a real-valued function, the constants $c_1$ and $c_2$ must be real and hence the constants $C_1$ and $C_2$ must constitute a complex-conjugate pair. Our two linearly independent real solutions of the homogeneous equation are
$$
    y_1(x)=e^{\alpha x}\cos(\beta x),\hspace{10mm} y_2(x)=e^{\alpha x}\sin(\beta x).
$$
Indeed, it is easy to check that these are linearly independent:
$$
    \frac{y_2(x)}{y_1(x)}=\tan(\beta x),
$$
which is non-constant.

**Example 10.6.** The differential equation governing *simple harmonic motion* is given by
$$
    x''(t)+\omega^2x(t)=0,
$$
where the constant $\omega$ is called the *angular frequency*. The above equation arises, for example, in the case of a mass $m$ on a frictionless surface, connected to a spring with spring constant $k$. In this case, $x(t)$ represents the position of the mass at time $t$, with $x=0$ corresponding to the equilibrium position, and
$$
    \omega=\sqrt{\frac{k}{m}}.
$$
If the spring-connected mass described here is pulled from position $x=0$ to position $x=A$ and then released from rest (at $t=0$), find the position of the mass at time $t$.
