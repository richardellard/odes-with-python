# Differential equations

A *differential equation* is an equation which relates an unknown function with its derivatives. Our goal is usually to find that unknown function which satisfies the equation.

**Example 6.1.** The equations
$$
    \frac{dx}{dt}=x^2+t^2
$$
and
$$
    \frac{d^2y}{dx^2}+3\frac{dy}{dx}+7y=0
$$
are differential equations.

**Definition.** An differential equation is called *ordinary* if it only involves differentiation with respect to a single variable.

**Definition.** An differential equation is called $n$*-th order* if the highest derivative appearing in the equation is an $n$-th order derivative.

Without additional conditions, in general, there will be many solutions to a differential equation; if possible, we generally want to find *all* solutions.

**Example 6.2.** The equation $\frac{dy}{dx}=2xy$ is satisfied by the function $y(x)=Ce^{x^2}$ for any constant $C$. Furthermore, it can be shown that *any* solution to this equation must take this form (for some value of $C$).

Very often, a scientific principal or law can be translated into a differential equation. The solution to the differential equation provides insight into the behaviour of the system.

**Example 6.3.** **Newton's Law of Cooling** states that the rate of change of temperature $T$ of a body with respect to time $t$ is proportional to the difference between $T$ and the surrounding temperature $A$, that is
$$
    \frac{dT}{dt}=-k(T-A), \hspace{10mm} k>0.
$$
From the above equation, what happens when $T>A$ and what happens when $T<A$? Does this agree with your intuition about the physical process?

By solving the above differential equation, we are able to predict the temperature of the body at any time in the future.

**Example 6.4.** The size of a population $P$ with constant birth and death rates is governed by the equation
$$
    \frac{dP}{dt}=kP
$$
(here, $k$ is the difference between the birth and death rates). Any function of the form $P(t)=Ce^{kt}$ is a solution (check this). Here, we again see infinitely many solutions - this is typical of differential equations. Suppose, in particular, we are considering the population $P$ of a colony of bacteria at time $t$, where $t$ is measured in hours from some initial time $t=0$. If the initial population $P(0)=1000$, and the population doubles every hour, find the constants $C$ and $k$ and hence find the function which describes the size of the colony at any time $t$.

**Note.** The above equation and its solution also arise in the context of continuously compounded interest. If a sum of money invested under compound interest is denoted $S(t)$, then the growth of that sum of money over a time step $\Delta t$ is governed by the equation
$$
    \Delta S=rS\Delta t,
$$
where $r$ is the interest rate. Taking the limit as $\Delta t\rightarrow0$,
$$
    \frac{dS}{dt}=\lim_{\Delta t\rightarrow0}\frac{\Delta S}{\Delta t}=rS.
$$

**Definition.** The *general solution* of an ordinary differential equation is the (generally infinite) family of functions which satisfy the equation.

**Definition.** An *initial condition* specifies the value of the solution at a particular point.

**Definition.** A *particular solution* is one that satisfies both the differential equation and some initial condition(s).

**Definition.** An *initial value problem* consists of a differential equation together with one or more initial conditions.

The following picture illustrates the family of solutions to Example 6.4 (the value $k=\ln(2)$ is fixed by the condition that the population doubles every hour).

```python
import numpy as np
from matplotlib import pyplot as plt
# make sure the plot is displayed in the notebook
%matplotlib inline

k = np.log(2)
C_values = [-12, -6, -3, -1, -0.5, 0.5, 1, 3, 6, 12]

t_range = (-2, 3)
P_range = (-8, 8)

# visual flurishes
fig, ax = plt.subplots()
fig.set_size_inches(6, 6)
ax.set_aspect(5/16)
plt.title('General solution for $k = \ln(2)$', y = 1.1)
plt.grid(True)
plt.xlabel('$t$', labelpad=25)
plt.ylabel('$P$').set_rotation(0) # prevents vertical text

def P(t, C):
    return C * np.exp(k*t)

t = np.arange(t_range[0], t_range[1], 0.01)

for C in C_values:
    plt.plot(t, P(t, C))

# set limits of the plot
plt.xlim(t_range) # plt.xlim(P_range[0],P_range[1])
plt.ylim(P_range) # plt.ylim(P_range[0],P_range[1])

# label the curves ###################################################################
def t_pos(C):
    return np.log(P_range[1]/C)/k

space = 1
ax.text(t_pos(12), P_range[0]-space, '$C=-12$', horizontalalignment='center',
        verticalalignment='top')
ax.text(t_pos(6), P_range[0]-space, '$C=-6$', horizontalalignment='center',
        verticalalignment='top')
ax.text(t_pos(3), P_range[0]-space, '$C=-3$', horizontalalignment='center',
        verticalalignment='top')
ax.text(t_pos(1), P_range[0]-space, '$C=-1$', horizontalalignment='center',
        verticalalignment='top')

space = 0.5
ax.text(t_pos(12), P_range[1]+space, '$C=12$', horizontalalignment='center',
        verticalalignment='bottom')
ax.text(t_pos(6), P_range[1]+space, '$C=6$', horizontalalignment='center',
        verticalalignment='bottom')
ax.text(t_pos(3), P_range[1]+space, '$C=3$', horizontalalignment='center',
        verticalalignment='bottom')
ax.text(t_pos(1), P_range[1]+space, '$C=1$', horizontalalignment='center',
        verticalalignment='bottom')

space = 0.25
ax.text(t_range[1]+space, P(t_range[1], 0.5), '$C=1/2$', horizontalalignment='left',
        verticalalignment='center')
ax.text(t_range[1]+space, P(t_range[1], -0.5), '$C=-1/2$', horizontalalignment='left',
        verticalalignment='center')
######################################################################################

plt.show()
```

In the above example, we required one parameter (one initial condition) to define the particular solution. In general, for an $n$-th order differential equation, we require $n$ parameters.
