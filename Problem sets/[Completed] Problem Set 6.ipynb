# Problem one

Use Theorem 7.1 to show that the differential equation $xy'=2y$ has a unique solution near any point in the $(x,y)$-plane where $x\neq0$.

**Answer:** We need only apply Theorem 7.1 with $f(x,y)=2y/x$, $\partial f/\partial y=2/x$.

# Problem two

Verify that the initial value problem
$$
    y'=2\sqrt{y}, \hspace{10mm} y(0)=0
$$
violates the conditions of Theorem 7.1. By plotting the slope field in Python for $x,y\in[0,1]$, can you see intuitively why there are two distinct solutions?

**Answer:** We see that
$$
    \frac{\partial f}{\partial y}=\frac{1}{\sqrt{y}},
$$
which is not continuous (or even defined) at $(0,0)$. Hence the conditions of Theorem 7.1 do not hold. From the slope field, we can identify two different solutions, both of which satisfy the initial condition:

```python
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def f(x,y):
    return 2 * math.sqrt(y)

x_range = (0, 1)
y_range = (0, 1)
density = 16 # width and height divided into this number of points

width = x_range[1] - x_range[0]
height = y_range[1] - y_range[0]
aspect_ratio = width/height

# decide where to draw the line segments
xs = np.linspace(x_range[0], x_range[1], density)
ys = np.linspace(y_range[0], y_range[1], density)

# decide the length of each line segment
# as a fraction of the total width (or height)
d = 1/density/2

fig, ax = plt.subplots()
ax.set_aspect(aspect_ratio)
plt.title('Slope field')
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$y$').set_rotation(0) # prevents vertical text

for x in xs:
    for y in ys:
        try:
            m = f(x, y)
        
            # length of each line segment computed so as to make
            # all segments visually of equal length
            l = d / math.sqrt((1/width**2+m**2/height**2)/(1+m**2))
        
            x0 = x - l * math.sqrt(1/(1+m**2)) / 2
            x1 = x + l * math.sqrt(1/(1+m**2)) / 2
            y0 = y + m * (x0 - x)
            y1 = y + m * (x1 - x)
            
            plt.plot([x0, x1], [y0, y1], color = 'black', linewidth = 0.5)
            
        except ZeroDivisionError:
            pass

h = 0.01
x = np.arange(x_range[0], x_range[1]+h, h)
for C in [0, 1]:
    y = C * x**2
    plt.plot(x, y)

plt.xlim(x_range)
plt.ylim(y_range)

plt.show()
```

# Problem three

A baseball traveling at moderate speed (say, less than 100 m s$^{-1}$) will experience a drag force which is roughly proportional to its velocity and in the opposite direction. If the ball is thrown directly downwards with some initial velocity, then the velocity of the ball at time $t$ can be described by the differential equation
$$
    \frac{dv}{dt} = -9.8-0.2v,
$$
where 9.8 m s$^{-2}$ is acceleration due to gravity and 0.2 is a reasonable value for the constant of proportionality. Use python to plot the slope field for this equation for $0\leq t \leq 25$ and $-100 \leq v\leq 0$, and from your plot, deduce an approximate value for the terminal velocity of the ball, that is, the velocity where the drag force exactly balances the force of gravity and the ball stops accelerating.

**Answer:** From the slope field, we see that the terminal velocity appears to be approximately -50 m s$^{-1}$:

```python
def f(t, v):
    return -9.8 - 0.2*v

x_range = (0, 25)
y_range = (-100, 0)
density = 16 # width and height divided into this number of points

width = x_range[1] - x_range[0]
height = y_range[1] - y_range[0]
aspect_ratio = width/height

# decide where to draw the line segments
xs = np.linspace(x_range[0], x_range[1], density)
ys = np.linspace(y_range[0], y_range[1], density)

# decide the length of each line segment
# as a fraction of the total width (or height)
d = 1/density/2

fig, ax = plt.subplots()
ax.set_aspect(aspect_ratio)
plt.title('Slope field')
plt.grid(True)
plt.xlabel('$t$')
plt.ylabel('$v(t)$').set_rotation(0) # prevents vertical text

for x in xs:
    for y in ys:
        try:
            m = f(x, y)
        
            # length of each line segment computed so as to make
            # all segments visually of equal length
            l = d / math.sqrt((1/width**2+m**2/height**2)/(1+m**2))
        
            x0 = x - l * math.sqrt(1/(1+m**2)) / 2
            x1 = x + l * math.sqrt(1/(1+m**2)) / 2
            y0 = y + m * (x0 - x)
            y1 = y + m * (x1 - x)
            
            plt.plot([x0, x1], [y0, y1], color = 'black', linewidth = 0.5)
            
        except ZeroDivisionError:
            pass

plt.show()
```

# Problem four

Use python to plot the slope field for the equation $y'=x-y$. From looking at the slope field, can you spot a particular solution to the equation *(hint: look for where the slope field appears to be moving in a straight line)*? Check that your solution is correct by differentiation. What (approximately) is the minimum value of the particular solution $y(x)$ satisfying $y(-4)=4$? What do you think the general solution looks like as $x\rightarrow\infty$?

**Answer:** From the slope field, we notice that $y(x)=x-1$ is one solution curve, a fact which is easily verified by differentiation. If we follow the slope field from the point $(-4,4)$, the resulting solution curve appears to reach a minimum value of -2 before it begins increasing again. As $x\rightarrow\infty$, all solution curves seem to be tending toward the line $y(x)=x-1$.
